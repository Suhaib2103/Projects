```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Top 10 most Sixes score
top_10= data.nlargest(10,'Sixes')
top_10= top_10.sort_values(by='Sixes',ascending=True)

plt.plot(top_10['Batsman'],top_10['Sixes'],color='m',linestyle=':',marker='H')

plt.title('Top 10 Most Sixes of the Tornament')
plt.xticks(rotation=90)
plt.show()
```

```python
#  Top_10 bowlers Took most Wickets
top10_bowler = data.nlargest(10,'Wickets')
top10_bowler= top10_bowler.sort_values(by='Wickets', ascending=True)

plt.barh(top10_bowler['Bowler'],top10_bowler['Wickets'],color='m')

plt.xticks(rotation=360)
plt.title('Top_10 bowlers Took most Wickets')
plt.show()
```

```python
#Team Win% in Heatmap
# Count total matches each team played
matches = pd.melt(data, value_vars=['Home', 'Away'], value_name='team')
total_matches = matches['team'].value_counts().reset_index()
total_matches.columns = ['Team', 'Matches']
print(total_matches)
```

```python
# Count wins for each team
wins = data['match_winner'].value_counts().reset_index()
wins.columns = ['Team', 'Wins']
wins
```

```python
# Merge and calculate win %
datas = pd.merge(total_matches, wins, on='Team', how='left').fillna(0)
datas['Win%'] = round((datas['Wins'] / datas['Matches']) * 100, 2)
datas
```

```python
# Plot heatmap
sns.heatmap(datas.set_index('Team')[['Win%']], annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("IPL Team Win Percentage")
plt.show()
```
